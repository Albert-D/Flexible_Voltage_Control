```python
import plotly.graph_objects as go
import numpy as np
from plotly.subplots import make_subplots

# 数据
methods = ['Linear', 'Safe-DDPG', 'RLC-FT']
metrics = ['Voltage Recovery Time', 'Reactive Power', 'Total Objective Cost']

# 数据值 (Mean ± Std)
data = {
    'Voltage Recovery Time': {
        'means': [38.69, 23.74, 8.19],
        'stds': [30.62, 22.76, 8.21],
        'unit': 'Steps'
    },
    'Reactive Power': {
        'means': [172.16, 113.28, 31.09],
        'stds': [289.74, 201.52, 56.36],
        'unit': 'MVar'
    },
    'Total Objective Cost': {
        'means': [542.07, 333.97, 140.28],
        'stds': [530.03, 384.29, 151.43],
        'unit': ''
    }
}

# Nature风格的颜色 - 高对比度，色盲友好
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # 蓝色、橙色、绿色

# 创建子图
fig = make_subplots(
    rows=1, cols=3,
    subplot_titles=[f"{metric}<br>({data[metric]['unit']})" for metric in metrics],
    horizontal_spacing=0.15
)


# 为每个指标创建分组柱状图
for i, metric in enumerate(metrics):
    means = data[metric]['means']
    stds = data[metric]['stds']
    
    for j, method in enumerate(methods):
        fig.add_trace(
            go.Bar(
                x=[method],
                y=[means[j]],
                error_y=dict(
                    type='data',
                    array=[stds[j]/2],
                    visible=True,
                    thickness=2,
                    width=3
                ),
                name=method,
                marker_color=colors[j],
                marker_line=dict(width=0.5, color='black'),
                showlegend=(i == 0),  # 只在第一个子图显示图例
                width=0.6
            ),
            row=1, col=i+1
        )

# 更新布局以符合Nature风格
fig.update_layout(
    # 整体布局
    width=1000,
    height=400,
    font=dict(
        family="Arial, sans-serif",
        size=12,
        color="black"
    ),
    
    # 背景
    plot_bgcolor='white',
    paper_bgcolor='white',
    
    # 图例
    legend=dict(
        x=1.02,
        y=1,
        xanchor='left',
        yanchor='top',
        bgcolor='rgba(255,255,255,0)',
        bordercolor='black',
        borderwidth=1,
        font=dict(size=11)
    ),
    
    # 标题
    title=dict(
        text="Performance Comparison of Different Controllers in 56-bus System",
        x=0.5,
        xanchor='center',
        font=dict(size=14, color='black')
    ),
    
    # 边距
    margin=dict(l=60, r=120, t=80, b=60)
)

# 更新x轴和y轴
for i in range(1, 4):
    fig.update_xaxes(
        row=1, col=i,
        showgrid=False,
        showline=True,
        linewidth=1,
        linecolor='black',
        tickfont=dict(size=10),
        title_font=dict(size=11)
    )
    
    fig.update_yaxes(
        row=1, col=i,
        showgrid=True,
        gridwidth=0.5,
        gridcolor='lightgray',
        showline=True,
        linewidth=1,
        linecolor='black',
        tickfont=dict(size=10),
        title_font=dict(size=11)
    )

# 更新子图标题格式
for annotation in fig['layout']['annotations']:
    annotation['font'] = dict(size=12, color='black')

# 显示图形
fig.show()

# 如果要保存为高质量图片（适合期刊投稿）
# fig.write_image("performance_comparison.png", width=1000, height=400, scale=3)
# fig.write_image("performance_comparison.pdf", width=1000, height=400)
```
